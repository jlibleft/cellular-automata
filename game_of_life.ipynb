```python
def change(board):
    # Tworzymy głęboką kopię, aby stare dane nie uciekły w trakcie obliczeń
    old_board = board.copy()
    rows, cols = board.shape
    
    for y in range(rows):
        for x in range(cols):
            # Liczenie sąsiadów (zakres -1 do 1 wokół komórki)
            # Używamy wycinania (slicing) dla szybkości w NumPy
            neighbors = old_board[max(0, y-1):min(rows, y+2), 
                                  max(0, x-1):min(cols, x+2)]
            count = np.sum(neighbors) - old_board[y, x]

            # Zasady Gry w Życie
            if old_board[y, x] == 1:
                if count < 2 or count > 3:
                    board[y, x] = 0
                else:
                    board[y, x] = 1
            if old_board[y, x] == 0:
                if count == 3:
                    board[y,x]=1
    return board
```

```python
def rescale(babay_board):
    board = np.zeros((500,500))
    for y in range(10):
        for x in range(10):
            if baby_board[y][x]==1:
                board[50*y:50*(y+1),50*x:50*(x+1)]=1
    return board    
```

```python
pygame.init()
grid_x, grid_y = 500,500
board = np.zeros(((grid_x,grid_y)))
screen = pygame.display.set_mode((grid_x,grid_y))
clock = pygame.time.Clock()
running = True
interval = 1300
next_toggle = pygame.time.get_ticks() + interval
first_time = True
baby_board = np.zeros((10,10))


while running:
    frame = pygame.surfarray.make_surface(board * 255)
    current_time = pygame.time.get_ticks()
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    if first_time:
        for i in range(25):
            baby_board[c_x[i]][c_y[i]]=1
        board = rescale(baby_board)
        screen.blit(frame, (0, 0))
        pygame.display.flip()
        first_time = False

    if current_time >= next_toggle:
        baby_board = change(baby_board)
        board = rescale(baby_board)
        next_toggle = current_time + interval



    
    for y in range(0,500,50):
        pygame.draw.line(frame, (60, 179, 113), [0,y], [grid_y, y], 1)
    for x in range(0,500,50):
        pygame.draw.line(frame, (60, 179, 113), [x,0], [x, grid_x], 1)
    

    screen.blit(frame, (0, 0))
    pygame.display.flip()

    clock.tick(60) 

pygame.quit()
```
