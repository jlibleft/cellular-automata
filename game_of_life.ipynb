```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
import pygame
```

```python
def generate_coordiantes(n):
    return [rd.choices(range(0, 450, 50),k=n)]
c_x,c_y = generate_coordiantes(10), generate_coordiantes(10)
```

```python
pygame.init()
grid_x, grid_y = 500,500
cell_x,cell_y = 2,2
screen = pygame.display.set_mode((grid_x,grid_y))
clock = pygame.time.Clock()
running = True
interval = 1000
next_toggle = pygame.time.get_ticks() + interval
board = np.zeros(((grid_x,grid_y)))
show_rectangle = True

while running:
    current_time = pygame.time.get_ticks()
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    
    if current_time >= next_toggle:
        show_rectangle = not show_rectangle
        next_toggle = current_time + interval
        print(f"dupa {show_rectangle}")
    board.fill(0)
    if show_rectangle:
        board[250:300,350:400] += 1
        board[50:100,150:200] += 1

    frame = pygame.surfarray.make_surface(board * 255)
    for y in range(0,500,50):
        pygame.draw.line(frame, (60, 179, 113), [0,y], [grid_y, y], 1)
    for x in range(0,500,50):
        pygame.draw.line(frame, (60, 179, 113), [x,0], [x, grid_x], 1)
    

    screen.blit(frame, (0, 0))
    pygame.display.flip()

    clock.tick(60) 

pygame.quit()
```
